```python
# Extract product title
title_tag = soup.find("span", {"class": "VU-ZEz"})  # Flipkart's title class
title = title_tag.get_text(strip=True) if title_tag else "Title not available"
```

```python
# Extract product price
price_tag = soup.find("div", {"class": "Nx9bqj CxhGGd"})  # Flipkart's price class
price = price_tag.get_text(strip=True) if price_tag else "Price not available"
```

```python
rating_tag = soup.find("div", {"class": "XQDdHH _6er70b"})  # Flipkart's rating class
rating = rating_tag.get_text(strip=True) if rating_tag else "No rating available"
```

```python
img_tag = soup.find("img", {"class": "_53J4C- utBuJY"})
```

```python
if img_tag and 'src' in img_tag.attrs:
    img_url = img_tag["src"]
     # Fetch the image content
    img_response = requests.get(img_url)
     # Display the image
    display(Image(img_response.content))
    print(f" Product Title: {title}")
    print(f" Price: {price}")
    print(f" Rating: {rating}\n")
else:
    print("Image not found.")
    
    
```
